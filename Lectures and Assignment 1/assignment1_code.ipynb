```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Adjusting the range of x1 (number of tables produced) for a reasonable limit
x1 = np.linspace(0, 1000, 400)

# Adjusting the constraints calculations to avoid division by zero or negative values under the square root
x2_1 = np.maximum((3600 - 6*x1) / 4, 0)
x2_2 = np.maximum((1600 - 4*x1) / 2, 0)
x2_objective = (38500 - 70*x1) / 50
# Creating the plot again
plt.figure(figsize=(8, 8))

# Plotting the constraints
plt.plot(x1, x2_1, label=r'(Constraint 1) $6x_1 + 4x_2 \leq 3600$ (Carpentry)', color="orange")
#plt.fill_between(x1, 0, x2_1, alpha=0.1, color='black')

plt.plot(x1, x2_2, label=r'(Constraint 2) $4x_1 + 2x_2 \leq 1600$ (Painting)', color="blue")
#plt.fill_between(x1, 0, x2_2, alpha=0.1, color='black')

plt.axhline(700, label=r'(Constraint 3) $x_2 \leq 700$ (Max chairs)', color="green")
#plt.fill_betweenx(x1, 0, 700, alpha=0.1, color='black')
plt.axvline(0, color="black")
plt.axhline(0, label=r'(Constraint 4) $x_1, x_2 \geq 0$ (Non-negativity)', color="black")
# Adding labels and legend

plt.plot(x1, x2_objective, label=r'(Objective Function) $70x_1 + 50x_2$', color="red")


plt.fill([0, 400, 50, 0], [0, 0, 700, 700], label="Feasible region", color='grey', alpha=0.2)
plt.scatter(50, 700, label=r'Optimal Solution $x_1=50, x_2=700$', color="black")

plt.xlim(-100, 1000)
plt.ylim(-100, 1000)
plt.xlabel(r'$x_1$ (Tables)')
plt.ylabel(r'$x_2$ (Chairs)')

# Adding grid lines
plt.grid(True, which='both')

# Add some labels and legend
plt.title('Constraints Visualization for Cosine Furniture Ltd')
plt.axhline(0, color='black', lw=2)
plt.axvline(0, color='black', lw=2)
plt.legend()

# Show plot
plt.show()
```
