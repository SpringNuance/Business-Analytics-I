```python
import numpy as np
import matplotlib.pyplot as plt

# Coefficients for each media
coefficients = {
    "TV": {"i": 1, "ai": -3000, "bi": 6000, "ci": 0.00003},
    "Podcasts": {"i": 2, "ai": -2500, "bi": 5000, "ci": 0.00005},
    "Radio": {"i": 3, "ai": -1500, "bi": 3000, "ci": 0.00005},
    "Social media": {"i": 4, "ai": -3000, "bi": 6000, "ci": 0.000035},
}

# Define the non-linear regression model
def customer_model(x, a, b, c):
    return a + b * (np.exp(c * x) / (1 + np.exp(c * x)))

# Generate a range of marketing expenditures for the visualization
x_values = np.linspace(0, 150000, 500)

# Plotting
plt.figure(figsize=(6, 4))

for media, coefs in coefficients.items():
    # Calculate the new customers for each media
    y_values = customer_model(x_values, coefs["ai"], coefs["bi"], coefs["ci"])
    plt.plot(x_values, y_values, label=f"{media}")

# Add labels and title
plt.xlabel('Marketing Expenditure (euros)')
plt.ylabel('New Customers')
plt.title('New Customers as a Function of Marketing Expenditure for Each Media')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
